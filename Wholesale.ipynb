```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Wholesale customers data.csv')
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
ws_df = df.drop(['Region'], axis = 1)
ws_df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
ws_df = scaler.fit_transform(ws_df)
#crime_df = pd.DataFrame(crime_df)
ws_df
```

```python
ws_df = pd.DataFrame(ws_df, columns = ['Channel', 'Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen'])
ws_df
```

```python
ws_df.isnull().sum()
```

```python
sns.pairplot(ws_df)
```

```python
x_columns = 7
x = ws_df.iloc[:, 1:]
x
```

```python
from sklearn.cluster import KMeans
scores = []
for i in range(1, 11):
    model = KMeans(n_clusters=i, random_state = 0)
    #model.fit(x.reshape(-1,1))
    model.fit(x)
    scores.append(model.inertia_)
    
plt.plot(range(1, 11), scores)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
```

```python
from sklearn.cluster import KMeans

kmeans5 = KMeans(n_clusters=5, tol=0.01, random_state=0)
#kmeans.fit(x.reshape(-1,1))
kmeans5.fit(x)
# clusters = kmeans.predict(x.reshape(-1,1))
clusters = kmeans5.predict(x)
print(clusters)
```

```python
centres5=kmeans5.cluster_centers_
print(centres5)
```

```python
#f, (ax1,ax2,ax3) = plt.subplots(nrows=1,ncols=3,figsize=(10,6))
colors = ['orange', 'blue', 'green', 'magenta', 'cyan','yellow']
plt.figure(figsize=(10,6))
for i in range(5):
    plt.scatter(x.iloc[clusters == i, 0], x.iloc[clusters == i, 1], c=colors[i])
    #plt.scatter(x[clusters == i], x[clusters == i], c=colors[i])
plt.scatter(kmeans5.cluster_centers_[:, 0], kmeans5.cluster_centers_[:, 1], color='red', marker='+', s=100)

plt.title('K-Means Clustering')
plt.xlabel('Category')
plt.ylabel('Spending')
plt.legend()
```

```python
import seaborn as sns
plt.figure(figsize=(15,7))
sns.scatterplot(x.iloc[clusters == 0, 0], x.iloc[clusters == 0, 1], color = 'yellow', label = 'Cluster 1',s=50)
sns.scatterplot(x.iloc[clusters == 1, 0], x.iloc[clusters == 1, 1], color = 'blue', label = 'Cluster 2',s=50)
sns.scatterplot(x.iloc[clusters == 2, 0], x.iloc[clusters == 2, 1], color = 'green', label = 'Cluster 3',s=50)
sns.scatterplot(x.iloc[clusters == 3, 0], x.iloc[clusters == 3, 1], color = 'grey', label = 'Cluster 4',s=50)
sns.scatterplot(x.iloc[clusters == 4, 0], x.iloc[clusters == 4, 1], color = 'red', label = 'Cluster 5',s=50)

sns.scatterplot(kmeans5.cluster_centers_[:, 0], kmeans5.cluster_centers_[:, 1], color = 'red', 
                label = 'Centroids',s=100,marker='+')
plt.grid(False)
plt.title('Clusters of customers')
plt.xlabel('Category')
plt.ylabel('Spending')
plt.legend()
plt.show()
```
